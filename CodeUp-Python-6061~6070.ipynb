[6065](https://www.codeup.kr/problem.php?id=6065). 3개의 정수(a, b, c)가 입력되었을 때, 짝수만 출력해보자. 3개의 정수(a, b, c)가 공백을 두고 입력된다.
0 ~ +2147483647 범위의 정수들이 입력되며 적어도 1개는 짝수이다. 짝수만 순서대로 줄을 바꿔 출력한다.

```python
n = map(int, input().split())
for x in n: # map 객체도 for문 가능(iterable)
  if x % 2 == 0:
    print(x)
```

[6067](https://www.codeup.kr/problem.php?id=6067). 0이 아닌 정수 1개가 입력되었을 때, 음(-)/양(+)과 짝(even)/홀(odd)을 구분해 분류해보자.

음수이면서 짝수이면, A

음수이면서 홀수이면, B

양수이면서 짝수이면, C

양수이면서 홀수이면, D

를 출력한다. 정수 1개가 입력된다. -2147483648 ~ +2147483647, 단 0은 입력되지 않는다.

```python
x = int(input())

if x < 0:
  if x % 2 == 0:
    print('A')
  else:
    print('B')
elif x > 0:
  if x % 2 == 0:
    print('C')
  else:
    print('D')
```

입력

-2147483648

출력

A
