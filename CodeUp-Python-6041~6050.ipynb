첫 번째 줄에 합,


두 번째 줄에 차,


세 번째 줄에 곱,


네 번째 줄에 몫,


다섯 번째 줄에 나머지,


여섯 번째 줄에 나눈 값을 순서대로 출력한다.
(실수, 소수점 이하 둘째 자리까지의 정확도로 출력)

```python
a, b = map(int, input().split())
# 합
print(a + b)
# 차
print(a - b)
# 곱
print(a * b)
# 몫
print(a // b)
# 나머지
print(a % b)
# 나누기
n = a / b
print(format(n, '.2f'))
```

10 3
