[6051](https://www.codeup.kr/problem.php?id=6051). 두 정수(a, b)를 입력받아
a의 값이 b의 값과 서로 다르면 True 를, 같으면 False 를 출력하는 프로그램을 작성해보자. 2개의 정수(a, b)가 공백을 두고 입력된다. -2147483647 <= a, b <= +2147483648

```python
a, b = map(int, input().split())
r= True if a != b else False
print(r)
```

입력

0 1

출력

True
